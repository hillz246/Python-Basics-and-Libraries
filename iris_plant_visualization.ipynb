**A histogram for 'sepal_length'**

```python
# A histogram for the 'sepal_length' column.
```

```python
#Import the Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('clean data/iris_plant_dataset.csv')
df
```

***A histogram for the 'sepal_length' column.***

```python
plt.figure(figsize=(8, 6))  # Set figure size
plt.hist(df['sepal_length'], edgecolor='black')

# Add labels and title
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# to understand histogram plot
df.describe()
```

***A bar chart showing the mean 'petal_width' for each species.***

```python
# Calculate the mean petal width for each species
mean_petal_width = df.groupby('class')['petal_width'].mean()

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(mean_petal_width.index, mean_petal_width.values, color=['pink', 'violet', 'purple'])

# Add labels and title
plt.xlabel('Species')
plt.ylabel('Mean Petal Width (cm)')
plt.title('Mean Petal Width for Each Iris Species')

# Show the plot
plt.show()
```

***A scatter plot comparing 'sepal_length' and 'sepal_width' for each***

```python
# Create a scatter plot with Seaborn
plt.figure(figsize=(8, 6))  # Set figure size
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='class', style='class', s=100)

# Add labels and title
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Show the plot
plt.show()
```
